```python
a = x.unsqueeze(2).expand(-1, -1, 5, -1)
a.shape
```

```python
a
```

```python
b = torch.empty(a.size())
for i in range(x.size(0)):
    points = x[i]
    for j, p in enumerate(points):
        dist = points.add(-p).pow(2).sum(dim=1)
        knn_indices = dist.topk(5, largest=False, sorted=False)[1]
        knn_points = points.gather(0, knn_indices.unsqueeze(-1).repeat(1,2))
        b[i][j] = knn_points - p
```

```python
c= torch.concat((a,b), dim=3)
c
```

```python
c.reshape(c.size(0), c.size(1)*c.size(2), c.size(-1)).shape
```
