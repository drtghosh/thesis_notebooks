```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import torch.nn as nn
import torchvision
import umap
from PIL import Image

import pytorch_metric_learning
import pytorch_metric_learning.utils.logging_presets as logging_presets
from pytorch_metric_learning import losses, miners, samplers, testers, trainers
from pytorch_metric_learning.utils.accuracy_calculator import AccuracyCalculator
```

```python
from create_data_dgcn_cc import Circ as cc
```

```python
data_instance = cc(0.4, 0.4, 0.2, 26)
fpc, fnc, pc, lp, tc, lt = data_instance.create_dataset(5, 2, [100], [0.4], 10)
```

```python
torch.manual_seed(26)
fpc, fnc, pc, lp, tc, lt = torch.tensor(fpc), torch.tensor(fnc), torch.tensor(pc), torch.tensor(lp), torch.tensor(tc), torch.tensor(lt)
```

```python
from models import MLPGrow
map_network = MLPGrow(h_nodes=16, num_layers=2, in_dim=2, out_dim=16)
```

```python
loss_func = losses.NPairsLoss()
```

```python
fpc.shape
```

```python
fpc = fpc[1]
fnc = fnc.reshape(-1, 2)
pl = torch.ones(fpc.size(0))
nl = -1 * torch.ones(fnc.size(0))
data = torch.concat((fpc, fnc), dim=0)
labels = torch.concat((pl, nl))
```

```python
num_epochs = 100
learning_rate = 0.1
optimizer = torch.optim.Adam([{'params': map_network.parameters()}], learning_rate)
training_loss = 0

for i in range(num_epochs):
    optimizer.zero_grad()
    embeddings = map_network(data)
    loss = loss_func(embeddings, labels)
    loss.backward()
    optimizer.step()
    training_loss = loss.item()
    print(f"Epoch:{i}, Loss: {training_loss}")
```
